# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
#create engine for sqlalchemy from file
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# view columns
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c["type"])
```

```python
# view columns
columns = inspector.get_columns('station')
for c in columns:
    print(c['name'], c["type"])
```

```python
# Save references to each table
Measurment = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
# calculate latest date- 2017-08-23
session.query(Measurment.date).order_by(Measurment.date.desc()).first()
#return the data
data =session.query(Measurment.date, Measurment.prcp).\
    filter(Measurment.date > '2016-08-23').\
    order_by(Measurment.date).all()
#Convert to Dataframe
df = pd.DataFrame(data, columns=['date', 'precipitation'])
df['date']= pd.to_datetime(df['date'])
#plot
df.plot.line(x= 'date', y='precipitation', figsize = (17,8), linewidth=2)
plt.xlabel('Date')
plt.ylabel('Precipition')
plt.title("Last 12 months of Precipitation")
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
session.query(Measurment.station, func.count(Measurment.station)).\
    group_by(Measurment.station).\
    order_by(func.count(Measurment.station).desc()).all()

# List the stations and the counts in descending order.
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
session.query(Measurment.station, func.min(Measurment.tobs), func.max(Measurment.tobs), func.avg(Measurment.tobs)).\
    filter(Measurment.station == 'USC00519281').all()
```

```python
# Choose the station with the highest number of temperature observations.
session.query(Measurment.station, func.count(Measurment.tobs)).\
    group_by(Measurment.station).\
    order_by(func.count(Measurment.tobs).desc()).all()
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
data = session.query(Measurment.tobs).\
        order_by(Measurment.date.desc()).\
        filter(Measurment.station == 'USC00519281').\
        filter(Measurment.date > '2016-08-23').all()
#return the data

#Convert to Dataframe
df = pd.DataFrame(data, columns=["temp"])
#plot
df.plot.hist()
plt.ylabel('Temperature')
plt.title("Temp for USC00519281 for the last 12 months")
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    # """TMIN, TAVG, and TMAX for a list of dates.
    
    # Args:
    #     start_date (string): A date string in the format %Y-%m-%d
    #     end_date (string): A date string in the format %Y-%m-%d
        
    # Returns:
    #     TMIN, TAVE, and TMAX
    
    # Query all passengers
    return session.query(func.min(Measurment.tobs), func.avg(Measurment.tobs), func.max(Measurment.tobs)).\
        filter(Measurment.date >= start_date).\
        filter(Measurment.date <= end_date).all()
                              
# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
        
    sel = [func.min(Measurment.tobs), func.avg(Measurment.tobs), func.max(Measurment.tobs)]
    return session.query(*sel).\
        filter(func.strftime("%m-%d", Measurment.date) == date).all()
    
daily_normals("01-01")
```

```python
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`

# Set the start and end date of the trip

# Use the start and end date to create a range of dates

# Stip off the year and save a list of %m-%d strings

# Loop through the list of %m-%d strings and calculate the normals for each date
start_date = "2016-07-15"
end_date = "2016-07-29"
dates = pd.date_range(start_date, end_date, freq='D').strftime("%m-%d")

vacay_min = []
vacay_avg = []
vacay_max = []

for day in dates:
    norms_dict = {}
    
    data = session.query(func.min(Measurment.tobs)).\
    filter(func.strftime("%m-%d", Measurment.date) == day).all()    
    norms_dict = data
    vacay_min.append(norms_dict)

    data = session.query(func.avg(Measurment.tobs)).\
    filter(func.strftime("%m-%d", Measurment.date) == day).all()    
    norms_dict = data
    vacay_avg.append(norms_dict)

    data = session.query(func.max(Measurment.tobs)).\
    filter(func.strftime("%m-%d", Measurment.date) == day).all()    
    norms_dict = data
    vacay_max.append(norms_dict)

vacay_min
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index

vacay_df = pd.DataFrame(index = pd.date_range(start_date, end_date, freq='D'))
vacay_df['tmin'] = np.ravel(vacay_min)
vacay_df['tavg'] = np.ravel(vacay_avg)
vacay_df['tmax'] = np.ravel(vacay_max)
vacay_df
```

```python
# Plot the daily normals as an area plot with `stacked=False`

vacay_df.plot(stacked=False)
```
